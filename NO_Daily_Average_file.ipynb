```python
import pandas as pd
import xarray as xr
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
```

```python
files=os.listdir('/directory/')    
path = r'/directory/' # use your path
all_files = glob.glob(path + "/*.file format")

li = []
lid=[]
liy=[]
for filename in all_files:
    dfl = pd.read_csv(filename, index_col=None, header=0)
    df = dfl[(dfl['NO_IRRF']>0.0) & (dfl['NO_IRRF']<5.0)]
    NO_ave= df[['NO_IRRF']].mean()
    dte=df[['DOY']].mean()
    yer=df[['YEAR']].mean()

    li.append(NO_ave)
    lid.append(dte)
    liy.append(yer)

df = pd.concat(li)
dtt=pd.concat(lid)
yrr=pd.concat(liy)

date= np.array(dtt)
year= np.array(yrr)

dff=pd.DataFrame(df)
dff.rename(columns={0:'NO_Day_ave'},inplace=True)

dff.insert(1, "DOY", date)
dff.insert(2,'YEAR',year)
#***********************************************************
day=dff['DOY']

year=int(yer)

out= pd.to_datetime(day-1, unit='D', origin=str(year))
dff.insert(0,'Date',out)
dfi=dff.sort_values(by=['DOY'])
dfi.to_csv('/directory.file format')
dfi
```
